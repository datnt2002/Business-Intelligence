```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)
y = 3*x**2 - 2* x+1

plt.plot(x, y, 'r')

plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 3x^2 - 2x + 1')

plt.show()
```
